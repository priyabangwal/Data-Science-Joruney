```python
from bs4 import BeautifulSoup
```

```python
import requests 
```

```python
flipkart = requests.get("https://www.flipkart.com/search?q=realme%20mobile")
print(flipkart.content)                                                            #To retrieve data
```

```python
type(flipkart.content)
```

```python
type(flipkart.text)
```

```python
soup = BeautifulSoup(flipkart.text, "html.parser" )               #html parser use to convert string on which type of object
print(soup.prettify)    #use to make content more easy
```

```python
soup.title
```

```python
soup.title.text
```

```python
soup.p
```

```python
soup.div
```

```python
soup.div.text
```

```python
mobile = soup.find("div", class_ ="_1AtVbE") #use to find div with class znxK_ 
mobile
```

```python
mob = soup.find_all("div", attrs={"class" : "_13oc-S"}) 
mob
```

```python
len(mob)
```

```python
mobile = mob[0]
name = mobile.find("div" , class_ = "_4rR01T")
name
```
