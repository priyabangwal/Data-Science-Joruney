```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker = "1")     #marker means type of points 
plt.show()
```

```python
plt.plot(x,y,marker = "o", linestyle="") #marker means type of points 
plt.show()
```

```python
plt.plot(x,y,marker = "v", linestyle ="--" , color="red")  #linestyle means style of line
plt.show()
```

```python
plt.plot(x,y,marker = "o", linestyle ="dashed", ms= 30, mfc= "red" , mec ="blue" )   #ms means marker size , mfc means marker face value , mec means marker edge color
plt.show()
```

```python
plt.plot(x,y,marker = "o", linestyle ="dashed", ms= 30, mfc= "red" , mec ="blue" , linewidth=3)
plt.show()
```

```python
#Scatter Plot
```

```python
plt.scatter([3,9,7],[100,200,300])
plt.show()
```

```python
plt.scatter([3,9,7],[100,200,300])
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.show()
```

```python
x = np.arange(5,25)
plt.scatter(x, x**2 , color ="r", label="x^2", marker="^" )
plt.scatter(x, x**3 , color ="g", label="x^3", marker="o" )
plt.xlabel('X axis')
plt.ylabel("Y axis")                                                      #legends means name of the labels
plt.legend()
plt.title("Scatter Plot")
plt.show()
```

##SubPlots

```python
plt.subplot(1,2,1)      # 1 row , 2 column , at 1
plt.plot([1,2,3], [7,8,9])
plt.title("Pepcoding")
plt.subplot(1,2,2)      # 1 row , 2 column, at 2
plt.plot([4,5,6], [7,8,9])
plt.title("Nados")
plt.show()
```

```python
plt.subplot(2,3,1)
x = np.round(np.random.rand(7)*7)
plt.scatter(x,y, s =[1000,1200,1300,400,1300,1200,1100], alpha = 0.4, color=['r','g','b','c', 'g','c','r'])

plt.subplot(2,3,3)
x = np.arange(5,25)
plt.scatter(x, x**2 , color ="r", label="x^2", marker="^" )
plt.scatter(x, x**3 , color ="g", label="x^3", marker="o" )
plt.xlabel('X axis')
plt.ylabel("Y axis")
plt.legend()
plt.title("Scatter Plot")
 
plt.subplot(2,3,5)
plt.scatter([3,9,7],[100,200,300])
plt.xlabel('X axis')
plt.ylabel("Y axis")

plt.subplot(2,3,4)
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y,marker = "o", linestyle ="dashed", ms= 30, mfc= "red" , mec ="blue" , linewidth=3)

plt.plot()
plt.show()
```

##Bar Chart

```python
plt.bar(['Jan','Feb','March','April','May'],[100,120,205,135,140], color =['r', 'b', 'c', 'g','y'], width =[0.4,0.6,0.7,0.5,0.8])
plt.xlabel('Months')
plt.ylabel('Study')
plt.show()
```

```python
plt.barh(['Jan','Feb','March','April','May'],[100,120,205,135,140], color =['r', 'b', 'c', 'g','y'], height =[0.4,0.6,0.7,0.5,0.8])
plt.xlabel('Months')
plt.ylabel('Study')
plt.show()
```

##Histogram - continous plot

```python
x = np.random.randn(10000)
plt.hist(x, bins = 1)
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x, bins = 1000)                                   #bins tells in how many parts we want to divide our graph
plt.show()
```

##Pie Chart

```python
plt.pie([7,8,10,12,5,2], labels =['CO2','O2','NO2','O','Mg','Others'], explode = [0.2,0.3,0.1,0.3,0.2,0.4], shadow = True) #explode meaning kitna piche hua
plt.show()
```

```python
df = px.data.iris()
df
```

```python
df.species.unique()                  #types of species
```

```python
setosa = df[df["species"] == "setosa"]    
virginica = df[df["species"] == "virginica"]
versicolor = df[df["species"] =="versicolor"]
```

```python
setosa.corr()                         #correlation tells the
```

```python
plt.scatter(setosa['petal_length'], setosa['petal_width'] , label = 'setosa')
plt.scatter(virginica['petal_length'], virginica['petal_width'] , label = 'virginica')
plt.scatter(versicolor['petal_length'], versicolor['petal_width'] , label = 'versicolor')
plt.legend()
plt.show()
```

```python
fig = px.scatter_3d(df , x = 'petal_length', y = 'petal_width' , z = 'sepal_width', color ='species')
fig.show()
```
