```python
import seaborn as sns
```

```python
tip = sns.load_dataset("tips")
tip
```

Rel Plot

```python
sns.relplot(x = 'total_bill',y='tip',data =tips)                        #similar like scatter plot
```

```python
sns.relplot(x = 'total_bill',y='tip',data =tips , hue='sex')  #to identify sex
```

```python
sns.relplot(x = 'total_bill',y='tip',data =tips, hue='sex', style='time')  #differntitae between time
```

```python
sns.relplot(x = 'total_bill',y='tip',data =tips, hue='sex', style='time', size='size')
```

```python
sns.relplot(x = 'total_bill',y='tip',data =tips, hue='sex', style='time', size='size', kind='line')
```

```python
sns.relplot(x = 'total_bill',y='tip',data =tips, hue='smoker', col='time')
```

Cat Plot

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm' )
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm' , hue='size')
```

```python
sns.catplot(x='day', y='total_bill', data=tips,  hue='smoker')
```

```python
sns.catplot(x='smoker', y='tip', data=tips,order=['No','Yes'])
```

BOX Plot

```python
sns.catplot(x='day' , y ='tip', kind='box',data=tips)
```

```python
sns.catplot(x='day' , y ='tip', kind='boxen',data=tips)
```

Violin Plot

```python
sns.catplot(x='total_bill' , y ='day', hue='sex', split =True , kind ='violin',data=tips)
```

```python
g =sns.catplot(y='total_bill' , x ='day', split =True , kind ='violin',data=tips, inner=None)
```

```python
sns.catplot(y='total_bill' , x ='day', split =True , kind ='violin',data=tips, inner=None)
sns.swarmplot(y='total_bill' , x ='day',data=tips , color='k', size=3 )
```
